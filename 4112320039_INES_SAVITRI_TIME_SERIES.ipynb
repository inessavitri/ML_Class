# Machine Learning for Time Series

Machine learning untuk time series adalah pendekatan yang menggunakan teknik-teknik machine learning untuk menganalisis dan memodelkan data yang berkaitan dengan urutan waktu. Data time series terdiri dari urutan observasi yang diambil dalam interval waktu yang teratur, seperti data keuangan harian, suhu harian, atau penjualan bulanan.

Pendekatan machine learning untuk time series dapat digunakan untuk berbagai tujuan, seperti prediksi nilai masa depan, deteksi anomali, pengelompokan, atau pengenalan pola. Pada umumnya, pendekatan machine learning untuk time series melibatkan pra-pemrosesan data, pemilihan fitur yang sesuai, pemilihan model yang tepat, dan evaluasi yang cermat terhadap performa model. Selain itu, teknik validasi silang dan pengujian yang sesuai juga diterapkan untuk memastikan generalisasi yang baik pada data baru.

## Regressor Method

Machine learning for time series dengan metode Regressor adalah pendekatan yang digunakan untuk memprediksi nilai-nilai berkelanjutan pada data time series menggunakan algoritma regresi. Dalam konteks ini, tujuan utamanya adalah untuk mengembangkan model yang dapat mempelajari pola-pola dan hubungan dalam data historis time series, dan kemudian menggunakan pengetahuan tersebut untuk melakukan prediksi pada data masa depan.

Metode Regressor dalam machine learning for time series dapat digunakan untuk berbagai tugas, seperti prediksi harga saham berdasarkan data historis, prediksi penjualan berdasarkan data penjualan sebelumnya, atau prediksi permintaan energi berdasarkan data konsumsi energi sebelumnya.

Beberapa metode Regressor yang umum digunakan dalam machine learning for time series adalah sebagai berikut:

1. Linear Regression: Metode ini menggunakan persamaan linear untuk memodelkan hubungan antara variabel input dengan variabel target. Regresi linear cocok untuk data dengan tren linier yang jelas.

2. Support Vector Regression (SVR): Metode ini merupakan varian dari Support Vector Machines (SVM) yang digunakan untuk masalah regresi. SVR mencoba membangun garis atau permukaan pemisah yang paling optimal antara data input dan target dengan meminimalkan kesalahan.

3. Random Forest Regression: Metode ini menggunakan ensemble learning dengan membangun banyak pohon keputusan (decision tree) dan menggabungkan hasil prediksi dari setiap pohon. Hal ini memungkinkan model untuk menangani pola yang kompleks dalam data time series.

4. Gradient Boosting Regression: Metode ini juga menggunakan ensemble learning dengan membangun beberapa pohon keputusan secara berurutan. Setiap pohon dibangun untuk memperbaiki kesalahan prediksi pohon sebelumnya, sehingga secara bertahap meningkatkan kualitas prediksi.

5. Long Short-Term Memory (LSTM): Metode ini merupakan jenis jaringan saraf rekursif (recurrent neural network/RNN) yang populer untuk memodelkan data time series. LSTM memiliki kemampuan untuk mengenali pola jangka panjang dan menjaga ingatan dari data sebelumnya, sehingga cocok untuk memprediksi time series yang kompleks.

Setiap metode Regressor memiliki kelebihan dan kekurangan masing-masing serta kondisi penggunaan yang sesuai. Pemilihan metode tergantung pada karakteristik data time series yang akan diprediksi dan tujuan analisis yang ingin dicapai.

## Application of the method in this analysis

### Extreme Gradient Boosting Regressor (XGBRegressor)

XGBRegressor adalah metode regresi yang menggunakan algoritma Extreme Gradient Boosting (XGBoost). XGBoost adalah sebuah teknik yang sangat populer dalam machine learning untuk memodelkan data dan melakukan prediksi, terutama pada tugas-tugas regresi dan klasifikasi.

XGBRegressor adalah implementasi regresi dari XGBoost, yang menggunakan ensemble learning dengan membangun banyak pohon keputusan (decision tree) secara berurutan. XGBoost memiliki beberapa fitur yang membuatnya efektif dalam memodelkan data time series, di antaranya:

1. Gradient Boosting: XGBoost menggunakan teknik gradient boosting untuk membangun pohon keputusan secara berurutan. Setiap pohon yang dibangun mencoba memperbaiki kesalahan prediksi model sebelumnya, sehingga secara bertahap meningkatkan kualitas prediksi.

2. Regularisasi: XGBoost menyertakan metode regularisasi yang membantu menghindari overfitting pada model. Regularisasi ini melibatkan parameter seperti max_depth (kedalaman maksimum pohon), min_child_weight (bobot minimal pada leaf), dan gamma (penalti untuk membagi node).

3. Feature Importance: XGBoost memberikan skor pentingnya fitur dalam mempengaruhi prediksi. Hal ini memungkinkan kita untuk mengevaluasi pengaruh setiap fitur dalam model dan mengidentifikasi fitur-fitur yang paling penting dalam memprediksi data time series.

4. Penanganan Data yang Hilang: XGBoost dapat menangani data yang hilang dengan sendirinya. Pada saat membangun pohon keputusan, XGBoost dapat menentukan apakah data yang hilang berada di sebelah kiri atau kanan pemisah, dan secara otomatis memperlakukan data yang hilang.

XGBRegressor memiliki banyak parameter yang dapat disesuaikan untuk meningkatkan kualitas model dan menyesuaikannya dengan karakteristik data time series yang spesifik. Beberapa parameter yang sering digunakan adalah n_estimators (jumlah pohon yang akan dibangun), learning_rate (tingkat pembelajaran), dan subsample (proporsi sampel yang akan digunakan dalam membangun setiap pohon). Dengan menggunakan XGBRegressor, Anda dapat memanfaatkan kekuatan algoritma XGBoost untuk memodelkan dan memprediksi data time series dengan akurasi yang tinggi.

### PJM Hourly Energy Consumption Data

PJM Interconnection LLC (PJM) adalah organisasi transmisi regional (RTO) di Amerika Serikat. Ini adalah bagian dari jaringan Interkoneksi Timur yang mengoperasikan sistem transmisi listrik yang melayani semua atau sebagian Delaware, Illinois, Indiana, Kentucky, Maryland, Michigan, New Jersey, Carolina Utara, Ohio, Pennsylvania, Tennessee, Virginia, Virginia Barat, dan Distrik dari Kolombia.

Hourly Energy Consumption Dataset atau Data konsumsi daya per jam berasal dari website PJM dan dalam satuan megawatt (MW) berisi kumpulan data yang mencatat konsumen energi listrik pada setiap jam dalam rentang waktu tertentu. Wilayah telah berubah selama bertahun-tahun sehingga data hanya dapat muncul untuk tanggal tertentu.

# Feature Creation

```python
pjme["hour"] = pjme.index.hour
pjme["day_of_week"] = pjme.index.day_of_week
pjme.head()
```

Keterangan

```python
pjme["quarter"] = pjme.index.quarter
pjme["month"] = pjme.index.month
pjme["year"] = pjme.index.year
pjme["day_of_year"] = pjme.index.day_of_year
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(x="year", y="PJME_MW", data=pjme)
plt.title("MW by year")
plt.show()
```

```python
train = pjme[pjme.index < "01-01-2015"]
test = pjme[pjme.index >= "01-01'2015"]
```

```python
x_train = train.drop("PJME_MW", axis=1)
x_test = test.drop("PJME_MW", axis=1)

y_train = train["PJME_MW"]
y_test = test["PJME_MW"]
```

# Model

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

model = XGBRegressor(n_estimators=1000, learning_rate=0.01, early_stopping_rounds=50)
model.fit(x_train, y_train, 
         eval_set=[(x_train, y_train) ,(x_test, y_test)], verbose=100)
```

Kita bisa melihat bahwa root mean squared error pada training set mulai turun dan juga pada test set turun juga.

Itu berhenti setelah 450 pohon, itu karena kesalahan root mean squared pada set validasi mulai memburuk setelah banyak pohon dibangun.

# Feature Importance

```python
importance_df = pd.DataFrame(data=model.feature_importances_, index=model.feature_names_in_, columns=["Importance"])
importance_df.sort_values("Importance")
```

```python
importance_df.sort_values("Importance").plot(kind="barh", title="Feature Importance")
plt.show()
```

# Prediction on Test Set

```python
test["Prediction"] = model.predict(x_test)
```

```python
pjme = pjme.merge(test["Prediction"], how="left", left_index=True, right_index=True)
pjme.head()
```

```python
ax = pjme["PJME_MW"].plot(figsize=(20,8))
pjme["Prediction"].plot(ax=ax, style=".")
plt.legend(["Truth Data","Prediction"])
plt.title("Raw Data and Prediction")
plt.show()
```

```python
ax=pjme.loc[(pjme.index > "01-01-2018") & (pjme.index < "01-08-2018")]["PJME_MW"].plot(figsize=(20,8), title="Week of Data")
pjme.loc[(pjme.index > "01-01-2018") & (pjme.index < "01-08-2018")]["Prediction"].plot(style=".")
plt.legend(["Truth Data", "Prediction"])
plt.show()
```

```python
model.score(x_test, y_test)
```

```python
score = np.sqrt(mean_squared_error(test["PJME_MW"], test["Prediction"]))
print(f"RMSE score on test set:{score:0.2f}")
```
