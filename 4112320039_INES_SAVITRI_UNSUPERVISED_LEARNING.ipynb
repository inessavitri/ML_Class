# Machine Learning

Machine Learning adalah bidang ilmu yang berfokus pada pengembangan algoritma yang dapat belajar dari dan membuat prediksi atau keputusan berdasarkan data. Proses ini melibatkan penggunaan metode matematika dan statistika yang diterapkan pada data untuk membangun model yang dapat meniru atau memprediksi perilaku yang diinginkan. ## Unsupervised Learning

Unsupervised Learning adalah metode Machine Learning yang tidak menggunakan data
berlabel. Model akan belajar dari data tanpa ada informasi mengenai target yang
diinginkan. Dalam Unsupervised Learning, algoritma akan mencoba mengidentifikasi pola
atau struktur yang tersembunyi dalam data. Unsupervised Learning memiliki banyak
potensi untuk digunakan dalam berbagai aplikasi karena dapat mengatasi masalah di mana
data berlabel tidak tersedia atau sulit diperoleh.

Unsupervised Learning umumnya mencakup dua jenis metode, yaitu Clustering dan
Dimensionality Reduction. Clustering adalah proses mengelompokkan data menjadi
beberapa kelompok atau cluster berdasarkan kesamaan atau perbedaan antara data
tersebut. Dimensionality Reduction adalah proses mengurangi jumlah fitur atau dimensi
dalam data tanpa kehilangan informasi yang penting. Teknik ini sangat berguna ketika
kamu memiliki data berdimensi tinggi yang sulit dianalisis atau ditampilkan secara visual

### Clustering

Metode unsupervised learning clustering adalah teknik yang digunakan dalam analisis data untuk mengelompokkan objek-objek dalam dataset ke dalam kelompok-kelompok yang serupa berdasarkan kesamaan atribut atau karakteristik tertentu. Dalam metode ini, tidak ada label atau informasi kelas yang tersedia untuk membimbing proses pengelompokkan. Tujuannya adalah menemukan struktur tersembunyi atau pola dalam data tanpa ada panduan eksternal.

Beberapa metode clustering yang populer termasuk:
1. K-Means: Metode ini membagi objek-objek dalam dataset ke dalam k kelompok yang berbeda dengan mencari pusat kelompok yang paling dekat dengan setiap objek. Setiap objek akan dikelompokkan ke dalam kelompok yang memiliki pusat terdekat.
2. Hierarchical Clustering: Metode ini membangun hierarki kelompok dengan menggabungkan atau memisahkan kelompok berdasarkan jarak atau kesamaan di antara objek. Hasilnya adalah dendrogram yang menggambarkan hubungan hierarkis antara kelompok.
3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Metode ini mengelompokkan objek berdasarkan kepadatan mereka dalam ruang atribut. Objek-objek yang cukup dekat secara spasial dan memiliki kepadatan yang cukup tinggi akan dianggap sebagai anggota kelompok, sedangkan objek yang lebih jarang akan dianggap sebagai noise.

Metode unsupervised clustering berguna dalam banyak aplikasi, seperti segmentasi pelanggan, analisis sosial media, pengelompokan data biomedis, dan pengelompokan pola pikir konsumen. Ini dapat membantu mengungkap pola-pola tersembunyi dalam data, mengidentifikasi kelompok-kelompok dengan karakteristik serupa, dan membantu pengambilan keputusan yang lebih baik.

### Dimensionality Reduction
Dimensionality reduction (pengurangan dimensi) adalah proses mengurangi jumlah atribut atau fitur dalam dataset sambil mempertahankan sebagian besar informasi yang relevan. Tujuan utama pengurangan dimensi adalah untuk mengatasi masalah kelebihan dimensi dan menghilangkan atribut yang tidak memberikan kontribusi signifikan terhadap pemahaman data.

Metode pengurangan dimensi dapat dikelompokkan menjadi dua kategori utama: metode linier dan metode non-linier.

Metode Linier:

1. Principal Component Analysis (PCA): Metode ini mengidentifikasi kombinasi linier baru dari atribut asli yang menggambarkan variasi terbesar dalam data. PCA menghasilkan komponen utama yang saling ortogonal satu sama lain dan dapat digunakan untuk mengurangi dimensi data.
2. Linear Discriminant Analysis (LDA): Metode ini mengidentifikasi kombinasi linier yang memaksimalkan pemisahan antara kelas dalam data. LDA sering digunakan dalam tugas klasifikasi.
3. Factor Analysis: Metode ini mencari faktor-faktor laten yang dapat menjelaskan korelasi dalam data. Faktor-faktor ini tidak secara langsung terlihat dan dapat digunakan sebagai atribut baru.


Metode Non-Linier:

1. t-Distributed Stochastic Neighbor Embedding (t-SNE): Metode ini memproyeksikan data ke dalam ruang dimensi yang lebih rendah sedemikian rupa sehingga menjaga jarak relatif antara objek yang mirip dalam data. t-SNE sering digunakan untuk visualisasi data yang kompleks.
2. Isomap: Metode ini memetakan data ke dalam ruang dimensi yang lebih rendah dengan mempertahankan struktur geometris data. Isomap memperhitungkan jarak geodesik antara objek dalam data.
3. Locally Linear Embedding (LLE): Metode ini mempelajari manifold yang terkandung dalam data dan memetakan data ke dalam ruang dimensi yang lebih rendah sedemikian rupa sehingga mempertahankan hubungan tetangga terdekat antara objek dalam manifold.

Pengurangan dimensi dapat membantu dalam menghilangkan kelebihan dimensi, mengurangi kompleksitas komputasi dan memperbaiki performa algoritma.

```python
def check_missing_value(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percentage = (data.isnull().sum() / data.isnull().count()*100).sort_values(ascending = True)
    return pd.concat([total, percentage], axis=1, keys = ['total', 'percentage'])
```

### K-Means

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters= i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df_scaler)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state=0)
kmeans.fit(df_scaler)
pred = kmeans.predict(df_scaler)
```

```python
#Building the model at 4 clusters, add the label to the dataframe
data['Cluster KMeans'] = pred
data['Cluster KMeans'].value_counts()
```
