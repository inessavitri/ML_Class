### **Logistic Regression**
Logistic   Regression   adalah   salah   satu   algoritma   dalam Machine Learning yang digunakan untuk klasifikasi data biner. Algoritma   ini   menggunakan   pendekatan   regresi   linier   untuk memodelkan hubungan antara variabel input dan output, dengan menghasilkan nilai probabilitas yang berkisar antara 0 dan 1. Dalam   hal   ini,   output   klasifikasi   dilakukan   berdasarkan probabilitas tersebut. Persamaan umum Logistic Regression yang bisa dikenal sebagai pembelajaran
supervised dalam termonilogi pembelajaran mesin jika labelnya diketahui.

\begin{equation*}
   \pi_{i}= \beta_{0} + \beta_{i}X_{i} +.....+\beta_{n}X_{n}
\end{equation*}

Terdapat tiga jenis utama logistic regression yang berbeda dalam eksekusi dan teori, yaitu:

1. Binary logistic regression,
2. multinomial logistic regression,
3. ordinal regresi logistik.

### **Decision Tree**
Model ini membangun sebuah pohon keputusan dari data training yang digunakan
untuk mengklasifikasikan rekaman CTG menjadi kategori risiko rendah,
sedang, atau tinggi. Decision Tree biasanya mudah dipahami dan dapat
memberikan hasil yang cukup akurat.
Terdapat banyak indeks untuk menghitung derajat impurity secara kuantitatif,
diantara indeks yang paling diketahui adalah

\begin{equation*}
    Gini : H(Q_{m})=\sum_{k} \rho_{mk}\left ( 1-\rho_{mk} \right )
    \end{equation*}
\begin{equation*}
    Entropy : H(Q_{m})= -\sum_{k} \rho_{mk}log\left ( \rho_{mk} \right )
\end{equation*}
\begin{equation*}
    Misclassification :H(Q_{m})= 1-max(\rho_{mk})
\end{equation*}

Setelah mendapatkan perbeaan derajat impurity, selanjutnya akan menghitung informasion gain\\
\begin{equation*}
    GAIN_{split} = Entropy(\rho )-\left ( \sum_{i=1}^{k}\frac{n_{i}}{n}Entropy(i) \right )
\end{equation*}

### **Random Forest**
Random forest atau Hutan Acak adalah perpanjangan dari pohon keputusan yang dikantongi. Sampel dataset pelatihan diambil dengan penggantian, tetapi pohon dibangun sedemikian rupa sehingga mengurangi korelasi antara masing-masing pengklasifikasi. Random Forest adalah perpanjangan dari bagging yang selain membangun pohon berdasarkan beberapa sampel data pelatihan juga membatasi fitur yang dapat digunakan untuk membangun pohon, memaksa pohon menjadi berbeda sehingga dapat meningkatkan kinerja.

Random forest melibatkan pembuatan sejumlah besar pohon keputusan dari sampel bootstrap dari kumpulan data pelatihan. Random forest juga melibatkan pemilihan subset fitur input (kolom atau variabel) pada setiap titik pemisahan dalam konstruksi pohon. Biasanya, membangun pohon keputusan melibatkan evaluasi nilai untuk setiap variabel masukan dalam data untuk memilih titik pisah. Dengan mereduksi fitur menjadi subset acak yang dapat dipertimbangkan pada setiap titik pemisahan, ini memaksa setiap pohon keputusan dalam ansambel menjadi lebih berbeda. Random forest disediakan melalui kelas Random Forest Regresor dan Random Forest Classifier. Kedua model beroperasi dengan cara yang sama dan mengambil argumen yang sama yang mempengaruhi bagaimana pohon keputusan dibuat.Keacakan digunakan dalam pembangunan model. Artinya setiap kali algoritma dijalankan pada data yang sama, maka akan menghasilkan model yang sedikit berbeda.

Prediksi untuk sampel yang tidak terlihat padat dibuat dengan membuat rata rata prediksi dari setiap pohon regresi\\
\begin{equation*}
    \hat{f}= \frac{1}{B} \sum_{b=1}^{B}f_{b} (){x}')
\end{equation*}

Selain itu, perkiraan ketidakpastian prediksi dapat dibuat sebagai standar deviasi prediksi dari setiap pohon regresi x\\
\begin{equation*}
    \sigma = \sqrt{\frac{\sum_{b=1}^{B}(f_{b}({x})'-f)^{2}}{B-1}}\\
\end{equation*}

### **Confusion matrix**

Confusion matrix digunakan untuk mengevaluasi performa model klasifikasi pada beberapa kelas kategori. Jika terdapat tiga kelas kategori, confusion matrix akan menjadi matriks persegi 3x3 dengan tiga baris dan tiga kolom.

Confusion matrix pada umumnya memiliki empat sel, yaitu True Positive (TP), True Negative (TN), False Positive (FP), dan False Negative (FN), yang mewakili jumlah prediksi yang benar dan yang salah dari model.

Berikut adalah contoh tampilan confusion matrix untuk tiga kelas kategori, yaitu A, B, dan C:

           A    B    C
       A  TP   FP   FN
       B  FP   TP   FN
       C  FP   FP   TP

- Diagonal utama dari kiri atas ke kanan bawah mewakili jumlah True Positive (TP) untuk setiap kelas kategori. Misalnya, pada posisi (A, A), itu adalah jumlah prediksi benar untuk kelas A (TP_A).
- Di luar diagonal utama, setiap sel mewakili jumlah False Positive (FP) atau False Negative (FN) untuk kombinasi kelas kategori tertentu. Misalnya, pada posisi (A, B), itu adalah jumlah prediksi salah (FP) untuk kelas B yang seharusnya A.

Untuk menganalisis confusion matrix, beberapa metrik evaluasi umum yang dapat dihitung dari nilai-nilai di dalamnya adalah sebagai berikut:

- Akurasi (Accuracy): Mencerminkan sejauh mana model mampu mengklasifikasikan dengan benar seluruh data, dihitung dengan (TP_A + TP_B + TP_C) / (Total data).
- Presisi (Precision): Mencerminkan sejauh mana prediksi positif model benar, dihitung dengan TP_A / (TP_A + FP_A), TP_B / (TP_B + FP_B), TP_C / (TP_C + FP_C) untuk masing-masing kelas.
- Recall (Sensitivitas atau True Positive Rate): Mencerminkan sejauh mana model mampu mengidentifikasi dengan benar data positif, dihitung dengan TP_A / (TP_A + FN_A), TP_B / (TP_B + FN_B), TP_C / (TP_C + FN_C) untuk masing-masing kelas.
- F1-Score: Menggabungkan presisi dan recall menjadi satu metrik, dihitung dengan 2 * (presisi * recall) / (presisi + recall) untuk masing-masing kelas.

Data Classification of The Cardiotocogram (CTG) Data for Fetal Health atau Klasifikasi Data Cardiotocogram (CTG) untuk Kesehatan Janin adalah suatu teknik analisis data yang digunakan untuk mengklasifikasikan rekaman CTG yang mengindikasi risiko kesehatan janin yanng lebih tinggi. - mean_value_of_long_term_variability : Nilai rata -rata dari variabilitas jangka panjang
- histogram_width : Lebar histogram yang dibuat menggunakan semua nilai dari suatu catatan
- histogram_min : Nilai minimum histogram
- histogram_max : Nilai maximum histogram
- histogram_number_of_peaks : Jumlah puncak dalam histogram ujian
- histogram_number_of_zeros : Jumlah nol dalam histogram ujian
- histogram_mode : Modus histogram
- histogram_mean : Rata-rata histogram
- histogram_median : Nilai tengah histogram
- histogram_variance : Varians histogram
- histogram_tendency : Kecenderungan histogram

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score


# Definisi models
knn = KNeighborsClassifier()
logreg = LogisticRegression()
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier()

# Melatih model dengan data training
knn.fit(X_train, y_train)
logreg.fit(X_train, y_train)
dtc.fit(X_train, y_train)
rfc.fit(X_train, y_train)

# Memprediksi data test
knn_pred = knn.predict(X_test)
logreg_pred = logreg.predict(X_test)
dtc_pred = dtc.predict(X_test)
rfc_pred = rfc.predict(X_test)

# Evaluasi metrik diberbagai model
hasil1 = [(accuracy_score(y_test, knn_pred) *100), (recall_score(y_test, knn_pred, average='weighted') *100), (f1_score(y_test, knn_pred, average='weighted') *100), (precision_score(y_test, knn_pred, average='weighted') *100)]
hasil2 = [(accuracy_score(y_test, logreg_pred) *100), (recall_score(y_test, logreg_pred, average='weighted') *100), (f1_score(y_test, logreg_pred, average='weighted') *100), (precision_score(y_test, logreg_pred, average='weighted') *100)]
hasil3 = [(accuracy_score(y_test, dtc_pred) *100), (recall_score(y_test, dtc_pred, average='weighted') *100), (f1_score(y_test, dtc_pred, average='weighted') *100), (precision_score(y_test, dtc_pred, average='weighted') *100)]
hasil4 = [(accuracy_score(y_test, rfc_pred) *100), (recall_score(y_test, rfc_pred, average='weighted') *100), (f1_score(y_test, rfc_pred, average='weighted') *100), (precision_score(y_test, rfc_pred, average='weighted') *100)]

dic = {'KNN' : [str(round((hasil1[0]))) + '%', str(round((hasil1[1]))) + '%', str(round((hasil1[2]))) + '%', str(round((hasil1[3]))) + '%'], 
       'Logistic Regression' : [str(round((hasil2[0]))) + '%', str(round((hasil2[1]))) + '%', str(round((hasil2[2]))) + '%', str(round((hasil2[3]))) + '%'],
       'Decision Tree' : [str(round((hasil3[0]))) + '%', str(round((hasil3[1]))) + '%', str(round((hasil3[2]))) + '%', str(round((hasil3[3]))) + '%'],
       'Random Forest' : [str(round((hasil4[0]))) + '%', str(round((hasil4[1]))) + '%', str(round((hasil4[2]))) + '%', str(round((hasil4[3]))) + '%']}

df_hasil = pd.DataFrame(data = dic, index=['Accuracy', 'Recall', 'Precision', 'F1-Score'])
df_hasil
```
